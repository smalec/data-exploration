```python
from sklearn import datasets
iris = datasets.load_iris() 

print 'Data: ', iris.data[:10]
print 'Target: ', iris.target
print 'Feature names: ', iris.feature_names
print 'Target names: ', iris.target_names
```

```python
iris_x, iris_y = iris.data[:, 0], iris.data[:, 1]
scatter(iris_x, iris_y, c='r')
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
```

```python
scatter(iris_x, iris_y, c='r')
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xlim(3, 9)
ylim(1, 5)
```

```python
scatter(iris_x, iris_y, c='r')
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xticks(range(3, 10))
yticks(range(1, 6))
```

```python
scatter(iris_x, iris_y, c=iris.target, cmap='spring')
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xticks(range(3, 10))
yticks(range(1, 6))
savefig("zadanie1.png")
```
