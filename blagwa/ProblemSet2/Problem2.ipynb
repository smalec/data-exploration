```python
%pylab inline

from sklearn import datasets
iris = datasets.load_iris() 

iris_x, iris_y = iris.data[:, 0], iris.data[:, 1]
```

```python
iris_x_no_versicolor = iris_x[iris.target != 1]
iris_y_no_versicolor = iris_y[iris.target != 1]

scatter(iris_x_no_versicolor, iris_y_no_versicolor, c=iris.target[iris.target != 1], cmap='spring')
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xticks(range(3, 10))
yticks(range(1, 6))
```

```python
x = np.array([4.75, 6.2])
line_params = (2, -8)

scatter(iris_x_no_versicolor, iris_y_no_versicolor, c=iris.target[iris.target != 2], cmap='spring')
plot(x, line_params[0]*x + line_params[1])
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xticks(range(3, 10))
yticks(range(1, 6))
```

```python
over_line = iris_y_no_versicolor > line_params[0] * iris_x_no_versicolor + line_params[1]
targets_no_versicolor = iris.target[iris.target != 1]
is_setosa = targets_no_versicolor == 0
hits = np.logical_not(np.logical_xor(over_line, is_setosa))
scatter(iris_x_no_versicolor, iris_y_no_versicolor, c=['lime' if hit else 'r' for hit in hits])
plot(x, line_params[0]*x + line_params[1])
title("Iris")
xlabel(iris.feature_names[0])
ylabel(iris.feature_names[1])
xticks(range(3, 10))
yticks(range(1, 6))
savefig("zadanie2a.png")
```
