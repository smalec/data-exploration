```python
%pylab inline

import numpy as np
```

```python
points = np.array([
        [1, 10],
        [2, 10],
        [3, 11],
        [4, 12],
        [5, 18],
        [6, 18],
        [7, 19],
        [8, 26],
        [9, 19],
        [10, 26],
    ])
x, y = points[:, 0], points[:, 1]
scatter(x, y)
```

```python
line_params = (2, 5)
ran = np.linspace(1, 11, 10)

scatter(x, y)
plot(ran, line_params[0]*ran + line_params[1], c='r')
savefig("zadanie3a.png")
```

```python
dists = np.abs(-y + line_params[0]*x + line_params[1])/sqrt(line_params[0]**2 + line_params[1]**2)
print sum(dists)
```

```python
line_params = (2.1, 5.5)

scatter(x, y)
plot(ran, line_params[0]*ran + line_params[1], c='r')

dists = np.abs(-y + line_params[0]*x + line_params[1])/sqrt(line_params[0]**2 + line_params[1]**2)
print sum(dists)

savefig("zadanie3b.png")
```
