```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
sepal_length_no_vsc = iris.data[:, 0][iris.target != 1]
sepal_width_no_vsc = iris.data[:, 1][iris.target != 1]
colors = iris.target[iris.target != 1]
```

```python
plt.scatter(sepal_length_no_vsc, sepal_width_no_vsc, c=colors, cmap="spring")
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
plt.scatter(sepal_length_no_vsc, sepal_width_no_vsc, c=colors, cmap="spring")
x = np.array([5.0, 6.2])
plt.plot(x, 2*x - 8)
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
is_above_line = sepal_width_no_vsc > 2 * sepal_length_no_vsc - 8
targets_no_vsc = iris.target[iris.target != 1]
is_setosa = targets_no_vsc == 0
is_virginica = targets_no_vsc == 2

is_setosa_above = np.logical_and(is_setosa, is_above_line)
is_virginica_below = np.logical_and(is_virginica, np.logical_not(is_above_line))

colors = np.logical_or(is_setosa_above, is_virginica_below)
colors = ['g' if c else 'r' for c in colors]

plt.scatter(sepal_length_no_vsc, sepal_width_no_vsc, c=colors)
x = np.array([5.0, 6.2])
plt.plot(x, 2*x - 8)
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.savefig("zadanie2a.png")
plt.show()
```

```python
a = 1.5
b = -4.8

is_above_line = sepal_width_no_vsc > a * sepal_length_no_vsc + b
targets_no_vsc = iris.target[iris.target != 1]
is_setosa = targets_no_vsc == 0
is_virginica = targets_no_vsc == 2

is_setosa_above = np.logical_and(is_setosa, is_above_line)
is_virginica_below = np.logical_and(is_virginica, np.logical_not(is_above_line))

colors = np.logical_or(is_setosa_above, is_virginica_below)
colors = ['g' if c else 'r' for c in colors]

plt.scatter(sepal_length_no_vsc, sepal_width_no_vsc, c=colors)
x = np.array([4.8, 6.2])
plt.plot(x, a*x + b)
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.savefig("zadanie2b.png")
plt.show()
```
