```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[1,10], [2,10], [3,11], [4,12], [5,18], [6,18], [7,19], [8,26], [9,19], [10,26]])
```

```python
plt.scatter(points[:, 0], points[:, 1])
plt.show()
```

```python
a = 2
b = 5
x = np.array([1.0, 11.0])

plt.scatter(points[:, 0], points[:, 1])
plt.plot(x, a*x + b, 'r')
plt.savefig("zadanie3a.png")
plt.show()
```

```python
a = 2
b = 5
dists = np.abs(a * points[:, 0] + b - points[:, 1]) / np.sqrt(a ** 2 + b ** 2)
np.sum(dists)
```

```python
a = 2.07
b = 5.8
dists = np.abs(a * points[:, 0] + b - points[:, 1]) / np.sqrt(a ** 2 + b ** 2)
print(np.sum(dists))

plt.scatter(points[:, 0], points[:, 1])
plt.plot(x, a*x + b, 'r')
plt.savefig("zadanie3b.png")
plt.show()
```
