```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_centres(K, d):
    angle = 2.0 * np.pi / K
    radius = (d / 2.0) / np.sin(angle / 2.0)
    centres = np.array([(np.sin(i*angle)*radius, np.cos(i*angle)*radius) for i in range(K)])
    return centres
    
def get_clouds(K, d):
    centres = get_centres(K, d)
    clouds = np.array([[np.random.normal(a_i, 1, 1000), np.random.normal(b_i, 1, 1000)] for (a_i, b_i) in centres])
    return clouds

def solve(K, d):
    clouds = get_clouds(K, d)
    clouds_x = clouds[:, 0].flatten()
    clouds_y = clouds[:, 1].flatten()
    colors = [[c] * 1000 for c in range(0, K)]
    plt.scatter(clouds_x, clouds_y, c=colors, cmap="spring")
    plt.show()
```

```python
solve(7, 5)
```

```python
solve(11, 10)
```

```python
solve(23, 15)
```
